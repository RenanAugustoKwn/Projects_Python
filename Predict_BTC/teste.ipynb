```python
import pandas as pd

df = pd.read_csv("PlanilhaBTC.CSV", sep=";")
print(df)
```

```python
df.plot(kind="scatter", x="ANO", y="DOLAR");
```

```python
from sklearn import linear_model
modelo = linear_model.LinearRegression()
x = df[["ANO"]]
y = df[["DOLAR"]]

modelo.fit(x, y)
```

```python
print(modelo.predict([[2012]])[0][0])
```
